### Изменим названия столбцов

Создадим функцию для приведения названий столбцов в нужный нам вид

````python
def col_norm(dataset):
    # Преобразуем имена столбцов в нижний регистр
    col_names = np.array(dataset.columns.str.lower())
    new_cols = []
    for col in col_names:
        col = col.replace(',%', "")
        col = "_".join(col.split(", ")).replace(',', "_") 
        # Разделяем имя столбца по пробелам и убираем лишние пробелы
        col = '_'.join([x.strip() for x in col.split()])     
        # Убираем все символы после открывающей скобки '('
        if "(" in col:
            col = col[:col.index('(')-1]
        new_cols.append(col)
    return new_cols
````

````python
data_farm.columns = col_norm(data_farm)
data_dad.columns = col_norm(data_dad)
data_buy.columns = col_norm(data_buy)
# Проверим что все хорошо
data_farm.info(), data_dad.info(), data_buy.info()
````

## Исследовательский анализ данных

### Обьединение данных

Обьединим данные из датафрейма ```data_dad``` с данными из```data_farm```

````python
data_merged = data_farm.merge(data_dad, on = 'id', how='outer')
data_merged.head()
````

### Статистический анализ признаков

````python
# Количественные признаки
kol_cols = []
# Категориальные переменные
cat_cols = []
for col in data_merged.columns[1:]:
    if is_numeric_dtype(data_merged[col]):
        kol_cols.append(col)
    else:
        cat_cols.append(col)
kol_cols, cat_cols
````

#### Количественные переменные

Создадим функцию для статистического анализа каждой количественной переменной

````python
def stat_kol(columns, data):
    for col in kol_cols:
        print(col)
        print(f"Мода {col} равна {data[col].mode()[0]}")
        print(f"Медиана {col} равна {data[col].median()}\n")
        print(f"Среднее арифметическое {col} равно {data[col].mean()}\n")
        print(f"Дисперсия {col} равна {data[col].var()}")
        print(f"Стандартное отклонение {col} равно {data[col].std()}\n")
        print(f"Размах {col} равен {data[col].max()-data[col].min()}")
        print(f"Межквартильный размах {col} - {st.iqr(data[col])}")
        
        # Визуализируем
        fig, axes = plt.subplots(1, 2, figsize=(12, 6))
        data[col].plot.box(ax=axes[0])
        axes[0].set_title(f'Диаграмма распределения {col}')
        axes[0].grid(True)
        axes[0].set_ylabel('Значение')
        data[col].plot.hist(bins=30, ax=axes[1])
        axes[1].set_title(f'Гистограмма {col}')
        axes[1].grid(True)
        axes[1].set_ylabel('Частота')
        axes[1].set_xlabel('Значение')
        plt.suptitle('')  # Убираем общий заголовок
        plt.show()
        print("\n-------------------")
````

````python
stat_kol(kol_cols, data_merged)
````

**Вывод**

+ В данных об удое есть выброс, который скорее всего явлется ошибкой и его нужно удалить, ведь такой большой удой невозможен
+ Почти все данные распоожены внутри "усов", но иногда встречаются более редкие значения, котроые выбиваются из всех остальных
+ Медиана почти у всех признаков находится в вехней части ящика, что говорит о том, что у большинства коров выскоие показатели
+ Некоторые количественные признаки, такие как жирность, спо, протеин, имеют скошенное распределение. Это указывает на вероятное наличие перекосов в данных. Скорее всего в данных существую подгруппы или данные зависят от категориальных призанков


#### Категориальные переменные

Создадим функцию для статистического анализа категориальных переменных

````python
def stat_cat(columns):
    for col in columns:
        print(f"Частотное распределение для {col}:")
        print(data_merged[col].value_counts())
        print("\nПроцентное соотношение:")
        print(data_merged[col].value_counts(normalize=True) * 100)
        
        # Визуализация распределения
        fig, axes = plt.subplots(1, 2, figsize=(12, 6))
        data_merged[col].value_counts().plot.bar(ax=axes[0])
        axes[0].set_title(f'Диаграмма распределения {col}')
        axes[0].grid(True)
        axes[0].set_ylabel('Значение')
        axes[0].tick_params(axis='x', rotation=90)
        data_merged[col].value_counts().plot.pie(
        autopct='%1.1f%%', 
        startangle=90, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 1}
        )
        plt.title(f'Распределение для {col}')
        plt.ylabel('')
        plt.show()
        
        print("\n-------------------")
````

````python
stat_cat(cat_cols)
````

**Вывод**

+ В категориальных признаках заметно неравномерное распределение классов, и в некоторых случаях, например в возрасте коров он является значительным

## Обучение модели линейной регрессии

Отберем признаки, которые подходят для линейной регрессии

Заметим, что характеристики молока никак не могут повлиять на удой, а являются его следствием, а значит следует исключить жирность, содержание белка и вкус молока

````python
data_lr = clean_data_farm.drop(['жирность', 'белок', 'вкус_молока'], axis=1)
````

### Первая модель

````python
X_1 = data_lr.drop('удой_кг', axis=1).drop('имя_папы',axis = 1)
y_1 = data_lr['удой_кг']

# разделение на тренировочную и тестовую выборки
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(
    X_1, 
    y_1, 
    random_state=42,
    test_size = 0.25
)
````

````python
# Выделим категориальные признаки
X_train_1_cat = X_train_1[[
    'порода',
  'тип_пастбища',
  'порода_папы_быка',
  'возраст'
]]
X_test_1_cat = X_test_1[[
    'порода',
  'тип_пастбища',
  'порода_папы_быка',
  'возраст'
]]
# выбор метода кодирования, его инициализация
encoder_1 = OneHotEncoder(drop='first', sparse_output=False)

# обучение кодирования на тренировочных данных
encoder_1.fit(X_train_1_cat)

# кодирование тренировочной выборки
X_train_1_ohe = encoder_1.transform(X_train_1_cat)

# кодирование тестовой выборки
X_test_1_ohe = encoder_1.transform(X_test_1_cat)

encoder_col_names = encoder_1.get_feature_names_out() 

# создайте датафрейм из закодированных данных
X_train_1_ohe = pd.DataFrame(X_train_1_ohe, columns=encoder_col_names)
X_test_1_ohe = pd.DataFrame(X_test_1_ohe, columns=encoder_col_names)
````

````python
# Выделим количественные признаки
X_train_1_col = X_train_1[['эке', 'спо']]
X_test_1_col = X_test_1[['эке', 'спо']]

scaler_1 = StandardScaler()

# обучите скейлер на количественных признаках тренировочной выборки
scaler_1.fit(X_train_1_col)

# масштабируйте тренировочную выборку
X_train_1_scaled = scaler_1.transform(X_train_1_col)

# масштабируйте тестовую выборку
X_test_1_scaled = scaler_1.transform(X_test_1_col)

X_train_1_scaled = pd.DataFrame(X_train_1_scaled, columns=['эке', 'спо'])
X_test_1_scaled = pd.DataFrame(X_test_1_scaled, columns=['эке', 'спо'])
````

````python
X_train_1 = pd.concat([X_train_1_ohe, X_train_1_scaled], axis=1)
X_test_1 = pd.concat([X_test_1_ohe, X_test_1_scaled], axis=1)
X_train_1.head(), X_test_1.head()
````

````python
# инициализируем модель линейной регрессии
model_lr_1  = LinearRegression()

# обучим модель на тренировочных данных
model_lr_1.fit(X_train_1, y_train_1)

# получим предсказания модели на тестовых данных 
predictions_1 = model_lr_1.predict(X_test_1)
````

````python
# Расчитаем метрики
r2_1 = r2_score(y_test_1, predictions_1)
mse_1 = mean_squared_error(y_test_1, predictions_1)
mae_1 = mean_absolute_error(y_test_1, predictions_1)
rmse_1 = mean_squared_error(y_test_1, predictions_1, squared = False)
print("R2_1 =", r2_1)
print("mse_1 =", mse_1)
print("mae_1 =", mae_1)
print("rmse_1 =", rmse_1)
````

````python
# проведем анализ остатков
residuals = y_test_1 - predictions_1

# постройте графики по количественным признакам — гистограмму частотности 
# распределения и диаграмму рассеяния
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].hist(residuals)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_xlabel('Остатки')

axes[1].scatter(x=predictions_1, y=residuals)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
````

+ Смещённое вправо нормальное распределение говорит о том, что модель недооценивает реальные значения целевого признака, её прогнозы меньше фактических показателей. 

+ А на диаграмме распределения дисперсии мы видим "рупор", а значит предсказания модели могут оказаться ненадежными

+ Значение коэффициента детерминации близко к единице — модель хорошо себя показывает в 78 процентах случаев

+ Есть куда улучшать модель, необходимы дополнительные исследования

Обучим модель логистической регрессии

````python
clean_data_farm['вкус_молока'] = clean_data_farm['вкус_молока'].apply(lambda x: 1 if x == 'вкусно' else 0)
````

````python
clean_data_farm
````

````python
X = clean_data_farm.drop(columns='вкус_молока').drop(columns='id')
y = clean_data_farm['вкус_молока']

RANDOM_STATE = 450


X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE
)

# Выделим категориальные признаки
X_train_cat = X_train[[
    'порода',
  'тип_пастбища',
  'порода_папы_быка',
  'возраст',
    'имя_папы'
]]
X_test_cat = X_test[[
    'порода',
  'тип_пастбища',
  'порода_папы_быка',
  'возраст',
    'имя_папы'
]]
# выбор метода кодирования, его инициализация
encoder = OneHotEncoder(drop='first', sparse_output=False)

# обучение кодирования на тренировочных данных
encoder.fit(X_train_cat)

# кодирование тренировочной выборки
X_train_ohe = encoder.transform(X_train_cat)

# кодирование тестовой выборки
X_test_ohe = encoder.transform(X_test_cat)

encoder_col_names = encoder.get_feature_names_out() 

# создайте датафрейм из закодированных данных
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)
# приводим значения всех признаков к единой шкале
X_train_col = X_train[['жирность', 'белок', 'эке', 'спо']]
X_test_col = X_test[['жирность', 'белок','эке', 'спо']]

scaler = StandardScaler()

# обучите скейлер на количественных признаках тренировочной выборки
scaler.fit(X_train_col)

# масштабируйте тренировочную выборку
X_train_scaled = scaler.transform(X_train_col)

# масштабируйте тестовую выборку
X_test_scaled = scaler.transform(X_test_col)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=['жирность', 'белок','эке', 'спо'])
X_test_scaled = pd.DataFrame(X_test_scaled, columns=['жирность', 'белок','эке', 'спо'])
X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
clf = LogisticRegression()

clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
````

````python
for feature_number, feature_name in enumerate(X_train.columns):
    weight = clf.coef_[0, feature_number]
    print(f"{feature_name:<20}: {weight:+.6f}") 
````

Заметим, что больше всех влияет на вкус молока `тип_пастбища`, а так же `имя_папы_быка`

````python
pd.DataFrame(zip(y_test, y_pred), columns = ['test','pred']).head() 
````

Построим матрицу ошибок

````python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
plt.ylabel('True label')
plt.xlabel('Predicted');  
````

Расчитаем метрики

````python
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
print("accuracy:", round(accuracy, 3))
print('recall: ',round(recall, 3))
print('precision: ',round(precision, 3))
````

### Анализ модели логистической регрессии

Т.к важно, чтобы фермер не получил невкусное молоко, то следует полностью исключить ошибку 1 рода, то есть FP (False Positive), а значит мы должны поучить precision 1.00

Посмотрим на порог классификации логистической регрессии

````python
y_proba = clf.predict_proba(X_test)[:,1]
data = pd.DataFrame(list(zip(y_test, y_proba)),
                    columns=['y_valid', 'y_proba']).sort_values(by='y_proba', ascending=False)

data.sample(5)
````

Посмотрим на значение метрики precision при разых порогах

````python
thresholds = [round(i,2) for i in np.linspace(0.1,1,num = 4,endpoint=False)]
thresholds.extend([0.73, 0.74, 0.75, 0.76, 0.77])
columns = []
for i in thresholds:
    columns.append('y_pred_'+str(i))
    data['y_pred_'+str(i)] = data['y_proba'].apply(lambda x: 1 if x >= i else 0)
results = []
for threshold in thresholds:
    # Применение порога
    y_pred = [1 if prob >= threshold else 0 for prob in y_proba]
    
    # Расчет метрик
    precision = precision_score(y_test, y_pred)
    
    # Добавление в таблицу
    results.append([f"y_pred_{threshold}", round(precision, 2)])

# Создание DataFrame для отображения результатов
results_df = pd.DataFrame(results, columns=['порог', 'precision'])

results_df.sort_values(by=['precision', 'порог'], ascending = False)
````

Заметим, что при пороге 0.77 мы получаем precision равное 1, а значит это значение мы и возьмем

````python
true_threshold = 0.77
y_pred = [1 if prob >= true_threshold else 0 for prob in y_proba]
````

Снова измерим метрики

````python
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
print("accuracy:", round(accuracy, 3))
print('recall: ',round(recall, 3))
print('precision: ',round(precision, 3))
````

Дополнительное улучшение качества возможно через обработку аномалий или кросс-валидацию

+ Важными метриками для экономических расчетов и анализа рисков являются precision и recall, а также метрики R² и RMSE для прогнозирования количественных показателей
