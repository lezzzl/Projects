## Обучение модели

Перенесите сюда код обучения модели и её результат вывода на экран.


(Код в этом разделе запускается в отдельном GPU-тренажёре, поэтому оформлен не как ячейка с кодом, а как код в текстовой ячейке)

```python

def load_train(path):
    data_csv = pd.read_csv(path + 'labels.csv')
    train_datagen = ImageDataGenerator(
        validation_split=0.25,
        rescale=1./255,
        horizontal_flip=True,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2
        )

    train_datagen_flow = train_datagen.flow_from_dataframe(
        dataframe=data_csv,
        directory=path + 'final_files/',
        x_col='file_name',
        y_col='real_age',
        target_size=(224, 224),
        batch_size=16,
        class_mode='raw',
        subset='training',
        seed=12345
        )

    return train_datagen_flow

def load_test(path):
    data_csv = pd.read_csv(path + 'labels.csv')
    test_datagen = ImageDataGenerator(
        validation_split=0.25,
        rescale=1./255
        )

    test_datagen_flow = test_datagen.flow_from_dataframe(
        dataframe=data_csv,
        directory=path + 'final_files/',
        x_col='file_name',
        y_col='real_age',
        target_size=(224, 224),
        batch_size=16,
        class_mode='raw',
        subset='validation',
        seed=12345
        )

    return test_datagen_flow

def create_model(input_shape):
    optimizer = Adam(lr=0.0001)

    backbone = ResNet50(input_shape=input_shape,
                        weights='/datasets/keras_models/resnet50_weights_tf_dim_ordering_tf_kernels_notop.h5',
                        include_top=False)

    model = Sequential()

    model.add(backbone)

    model.add(GlobalAveragePooling2D())

    model.add(Dense(units=1, activation='relu'))
    model.compile(optimizer=optimizer, loss='mean_squared_error',
                  metrics=['mae'])

    return model


def train_model(model, train_data, test_data, batch_size=None, epochs=20,
                steps_per_epoch=None, validation_steps=None):
    early_stopping = EarlyStopping(
        monitor='val_mae',    
        patience=2,            # кол-во эпох без улучшений до остановки
        restore_best_weights=True,  
        verbose=1
    )
    model.fit(
        train_data,
        validation_data=test_data,
        epochs=epochs,
        batch_size=batch_size,
        steps_per_epoch=steps_per_epoch,
        validation_steps=validation_steps,
        callbacks=[early_stopping],
        verbose=2
    )

    return model

```

## Анализ обученной модели

1. **Структура модели**:

Для оптимизации используется **Adam** с очень низким шагом обучения (0.0001), что может быть полезно для точной настройки модели.

2. **Процесс обучения**:  
   Во время обучения использовался **ранний стоппинг (EarlyStopping)**, который прекращает обучение, если не происходит улучшений в валидационной ошибке. В данном случае улучшение модели стабилизировалось на 5-й эпохе, и модель была сохранена с лучшими весами

### Рекомендации по улучшению качества модели

- **Улучшение предобработки данных**: обратить внимание на ориентацию лиц, устранить повёрнутые изображения или выровнять их с помощью алгоритмов распознавания лиц.

- **Расширение аугментации**: добавить больше разнообразия в обучение модели, включая повороты, изменения яркости, контраста, масштабирование и другие трансформации, чтобы повысить устойчивость к внешним условиям.

- **Финетюнинг** предобученной модели не только верхних слоёв, но и части ResNet50 — особенно на последних этапах обучения.

- **Сбор дополнительных данных**: если есть возможность — добавить больше изображений, особенно из возрастных групп, где наблюдаются наибольшие ошибки.

- **Использование регрессионной функции потерь, чувствительной к выбросам**, например, Huber loss, которая может быть более устойчива к нетипичным значениям возраста.

Эти шаги могут помочь снизить MAE и сделать модель более точной и полезной в реальной практике.
